```python
all_news_date = news_df.date
```

```python
nep_to_eng_numbers = {'०': '0', '१': '1', '२': '2', '३': '3','४': '4', '५': '5', '६': '6', '७': '7', '८': '8', '९': '9'}
```

```python
new_years = []
new_months = []
```

```python
for date in all_news_date:
    splitted = date.split(' ')
    year = splitted[-1]
    # month = month_map[splitted[1]]
    month = splitted[1]
    s = []
    for i in year:
        s.append(nep_to_eng_numbers[i])
    year = ''.join(s)
    new_years.append(int(year))
    new_months.append(month)
```

```python
new_months[-1]
```

```python
news_df['Year'] = new_years
news_df['Month'] = new_months
```
